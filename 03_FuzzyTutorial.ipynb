```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

# Generate synthetic 2D data (two clusters)
np.random.seed(42)
cluster1 = np.random.normal(loc=[2, 4], scale=[0.5, 0.5], size=(100, 2))
cluster2 = np.random.normal(loc=[8, 8], scale=[0.5, 0.5], size=(100, 2))

# Combine data
data = np.vstack((cluster1, cluster2))

# Transpose the data for clustering
data = data.T

# Perform fuzzy c-means clustering
n_clusters = 2
cntr, u, u0, d, jm, p, fpc = fuzz.cluster.cmeans(
    data, n_clusters, 2, error=0.005, maxiter=1000, init=None)

# Find the cluster memberships
cluster_membership = np.argmax(u, axis=0)

# Plot the results
plt.scatter(data[0, :], data[1, :], c=cluster_membership, cmap='coolwarm')
plt.scatter(cntr[:, 0], cntr[:, 1], marker='x', s=200, color='black', label='Centers')
plt.title('Fuzzy C-Means Clustering')
plt.legend()
plt.show()
```
